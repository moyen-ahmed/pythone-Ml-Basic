```python
batters=pd.read_csv('useing_dataset/batter.csv')
batters.head()  
```

```python
sample_df=batters.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.figure(figsize=(14, 9))
plt.scatter(sample_df['avg'], sample_df['strike_rate'],s=sample_df['runs'])


for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].iloc[i], sample_df['strike_rate'].iloc[i], sample_df['batter'].iloc[i], fontsize=9)
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.title('Batter Average vs Strike Rate')
plt.show()
```

# Horizontal and Vertical line

```python
plt.figure(figsize=(14, 9))
plt.scatter(sample_df['avg'], sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130, color='red', linestyle='--', label='Strike Rate Threshold')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].iloc[i], sample_df['strike_rate'].iloc[i], sample_df['batter'].iloc[i], fontsize=9)
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.title('Batter Average vs Strike Rate')
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')    
ax.set_title('Batter Average vs Strike Rate')
ax.axhline(130, color='blue', linestyle='--', label='Strike Rate Threshold')
ax.legend()
plt.show()
```

```python
fig,ax=plt.subplots(nrows=2, ncols=1, figsize=(10, 10))
ax[0].scatter(batters['avg'], batters['strike_rate'], color='red', marker='+')
ax[0].set_title('Batter Average vs Strike Rate')
ax[0].set_xlabel('Average')     
ax[0].set_ylabel('Strike Rate')
ax[1].scatter(batters['avg'], batters['runs'], color='green', marker='x')
ax[1].set_title('Batter Average vs Runs')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Runs')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
ax[0,0].scatter(batters['avg'], batters['strike_rate'], color='red', marker='+')
ax[0,0].set_title('Batter Average vs Strike Rate')
ax[0,0].set_xlabel('Average')
ax[0,0].set_ylabel('Strike Rate')
ax[0,1].scatter(batters['avg'], batters['runs'], color='green', marker='x')
ax[0,1].set_title('Batter Average vs Runs')
ax[0,1].set_xlabel('Average')
ax[0,1].set_ylabel('Runs')
ax[1,0].hist(batters['avg'], bins=20, color='blue', alpha=0.7)
ax[1,0].set_title('Distribution of Average')
ax[1,0].set_xlabel('Average')
ax[1,0].set_ylabel('Frequency')
ax[1,1].hist(batters['strike_rate'], bins=20, color='orange', alpha=0.7)
ax[1,1].set_title('Distribution of Strike Rate')
ax[1,1].set_xlabel('Strike Rate')
ax[1,1].set_ylabel('Frequency')
plt.suptitle('Batter Statistics Analysis', fontsize=16)
plt.tight_layout()
plt.show()
```

# 3D Surface Plots

```python
x=np.linspace(-10, 10, 100)
y=np.linspace(-10, 10, 100)
np.meshgrid(x, y)
#meshgrid is used to create a rectangular grid out of two given one-dimensional arrays representing the Cartesian indexing or Matrix indexing.
# It is often used in 3D plotting to create a grid of points over which a function can be evaluated.

X, Y = np.meshgrid(x, y)
X.shape, Y.shape
```

```python
Z=X**2 + Y**2 
Z.shape
```

```python
fig=plt.figure(figsize=(10, 8))
ax=fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot of Z = X^2 + Y^2')
fig.colorbar(ax.plot_surface(X, Y, Z, cmap='viridis'), ax=ax, shrink=0.5, aspect=5)
plt.show()
```
