```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = pd.read_csv('useing_dataset/iris.csv')
iris.head()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Iris Sepal vs Petal Length')
plt.show()
```

```python
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'], cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Iris Sepal vs Petal Length')
colorbar = plt.colorbar()
colorbar.set_ticks([0, 1, 2])
plt.show()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'], cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Iris Sepal vs Petal Length')
colormap = plt.get_cmap('viridis')
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap=colormap)
colorbar = plt.colorbar()
plt.show()
```
