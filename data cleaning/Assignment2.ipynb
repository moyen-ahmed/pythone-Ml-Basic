```python

df = pd.read_csv("using dataset/housing.csv", delim_whitespace=True)

# Rename the columns
df.columns = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD',
    'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'
]

# Plot the distribution of the target variable
plt.figure(figsize=(10, 6))
sns.histplot(df['MEDV'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of MEDV (Median Value of Homes)')
plt.xlabel('MEDV ($1000s)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Compute correlation matrix
correlation_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Correlation Matrix of Boston Housing Dataset")
plt.show()
```

```python
features = ['RM', 'LSTAT', 'PTRATIO', 'NOX']
plt.figure(figsize=(16, 10))

for idx, feature in enumerate(features):
    plt.subplot(2, 2, idx + 1)
    sns.scatterplot(x=df[feature], y=df['MEDV'], color='green')
    plt.title(f'MEDV vs {feature}')
    plt.xlabel(feature)
    plt.ylabel('MEDV')

plt.tight_layout()
plt.show()
```
